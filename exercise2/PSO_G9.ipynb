```python
import numpy as np
from sko.PSO import PSO
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def g9(x):
    return (x[0]-10)**2 + 5*(x[1]-12)**2 + x[2]**4 +3*(x[3]-11)**2 +10*x[4]**6 +7*x[5]**2 + x[6]**4 -4*x[5]*x[6] -10*x[5] -8*x[6]
```

```python
ueq = [
    lambda x: -127 + 2*x[0]**2 + 3*x[1]**4 + x[2] + 4*x[3]**2 + 5*x[4],
    lambda x: -282 + 7*x[0] + 3*x[1] + 10*x[2]**2 + x[3] - x[4],
    lambda x: -196 + 23*x[0] + x[1]**2 + 6*x[5]**2 - 8*x[6],
    lambda x: 4*x[0]**2 + x[1]**2 - 3*x[0]*x[1] + 2*x[2]**2 + 5*x[5] - 11*x[6]
]
```

```python
pso = PSO(func=g9, n_dim=7, pop=100, max_iter=1000, lb=[-10] * 7, ub=[10] * 7, constraint_ueq = ueq, w=0.8, c1=0.5, c2=0.5)
pso.run()
print('best_x is ', pso.gbest_x, 'best_y is', pso.gbest_y)
```

```python
finite_values = [iteration for iteration in pso.gbest_y_hist if not np.isinf(iteration).any()]

data = [value for iteration in finite_values for value in iteration]

plt.plot(data)
plt.show()
```
