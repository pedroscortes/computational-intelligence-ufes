```python
import numpy as np
from sko.PSO import PSO
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def f5(x):
    return np.sum([xi**2 - 10 * np.cos(2 * np.pi * xi) + 10 for i, xi in enumerate(x, 1) if i <= 30])
```

```python
pso = PSO(func=f5, n_dim=30, pop=100, max_iter=300, lb=[-5.12] * 30, ub=[5.12] * 30, w=0.8, c1=0.5, c2=0.5)
pso.run()
print('best_x is ', pso.gbest_x, 'best_y is', pso.gbest_y)
```

```python
plt.plot(pso.gbest_y_hist)
plt.show()
```
