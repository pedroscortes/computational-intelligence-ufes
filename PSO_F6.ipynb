```python
import numpy as np
from sko.PSO import PSO
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def f6(x):
  n = len(x)
  term_1 = np.sum(np.power(x, 2)) / 4000
  term_2 = np.prod(np.cos(x / np.sqrt(np.arange(1, n + 1))))
  return term_1 - term_2 + 1
```

```python
pso = PSO(func=f6, n_dim=30, pop=100, max_iter=1000, lb=[-600] * 30, ub=[600] * 30, w=0.8, c1=0.5, c2=0.5)
pso.run()
print('best_x is ', pso.gbest_x, 'best_y is', pso.gbest_y)
```
