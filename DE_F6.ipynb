```python
import numpy as np
from sko.DE import DE
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def f6(x):
  n = len(x)
  term_1 = np.sum(np.power(x, 2)) / 4000
  term_2 = np.prod(np.cos(x / np.sqrt(np.arange(1, n + 1))))
  return term_1 - term_2 + 1
```

```python
de = DE(func=f6, n_dim=30, size_pop=100, max_iter=500, lb=[-600] * 30, ub=[600] * 30)

best_x, best_y = de.run()
print('best_x:', best_x, '\n', 'best_y:', best_y)
```

```python
Y_history = pd.DataFrame(de.all_history_Y)
fig, ax = plt.subplots(2, 1)
ax[0].plot(Y_history.index, Y_history.values, '.', color='red')
Y_history.min(axis=1).cummin().plot(kind='line')
plt.show()
```
